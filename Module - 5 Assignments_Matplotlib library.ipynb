```python
#import pandas library for access the databse of 'TelcoCustomer'
import pandas as pd
#read the datset
df = pd.read_csv('TelcoCustomer.csv')
```

```python
#glance of dataset
df.head()
```

##  Solution 1: Build a bar plot for 'Contract' column

```python
#import library
import matplotlib.pyplot as plt
%matplotlib inline

#Prepare Dataset
x = list(set(df['Contract']))
y = list(pd.value_counts(df['Contract']))

#subplot
ax1 = plt.subplot()

#plot bar
ax1.bar(x,y, color='orange')

#set labels
plt.xlabel('Contract type of customer')
plt.ylabel('Count')

#title of the plot
plt.title('Distribution of Contract')

#show
plt.show()
```

## Solution 2: Build the histogram for 'MonthlyCharges' column

```python
#prepare dataset
x = df['MonthlyCharges']

#subplot 
ax1 = plt.subplot()

#plot histogram
ax1.hist(x, color='forestgreen')

#add axis labels
plt.xlabel('Monthly Charges Incured')
plt.ylabel('Count')

#add title
plt.title('Distribution of monthly charges')

#show
plt.show()
```

## Solution 3: Build a scatter plot between 'PaymentMethod' and 'MonthlyCharges'

```python
#add figure size
plt.figure(figsize=(20,10))

#preapre 
x = df.loc[0:50,'tenure']
y = df.loc[0:50, 'TotalCharges']

plt.scatter(x,y, color='Indigo')
plt.xlabel('Tenure of Customer')
plt.ylabel('Total Charges Incurred')
plt.title('Total Charges Vs Tenure')
plt.show()
```

## Solution 4: Build a box plot between 'Payment method' and 'Monthly Charges'

```python
#add figure size
plt.figure(figsize=(20,15))

#Prepare Dataset
levelPM = list(df['PaymentMethod'].unique())
xlist = []
for level in levelPM:
#     print(level)
    l = df[df['PaymentMethod']==level]
    l = list(l['MonthlyCharges'])
    xlist.append(l)

#Plot Box plot    
pl = plt.boxplot(xlist, showmeans=True, labels=levelPM, patch_artist=True)

# print(pl['boxes'])
for box in pl['boxes']:
    box.set_facecolor(color='olive')
    
#add labels   
plt.xlabel('Payment method of customer')
plt.ylabel('Monthly Charges Incurred')

#add title
plt.title('Monthly Charges Vs Payment Method')

#show
plt.show()
```
